```python
import json

file_path = "xx/tasks/description/merged_updated+combo.json"
with open(file_path,'r') as f:
    data = json.load(f)
```

```python
import random
all_data = []
questions = ["What do these images depict?",
       "What do you observe in these images?",
       "What information do these images convey?",
       "What diagnostic features are shown in the content of these images?",
       "What information is reflected in the details of these images?",
       "What key diagnostic information can be extracted from these images?",
       "What possible diagnostic indicators are revealed in these images?",
       "What pathological signs are present in the image content?",
       "What pathological characteristics are present in these images?",
       "What disease-related details do you observe in these images?"]
for dp in data[-20:]:
    for combo in ["combo_0","combo_1","combo_2","combo_3","combo_4"]:
        messages = {"id":dp['patient']}
        message = [
                        {
                            "role": "user",
                            "content": []
                        },
                        {
                            "role": "assistant",
                            "content": []
                        }
                    ]
        do_add = False
        if len(dp[combo])!=0:
            do_add = True
            for image in dp[combo]:
                message[0]["content"].append({"type": "image", "image": image})
        message[0]["content"].append({"type": "text", "text": random.choice(questions)})
        if do_add:
            message[1]["content"].append({"type": "text", "text":dp["Image_Description"]})
            messages["messages"] = message
            if "image_Diagnosis" in dp:
                message.append({
                            "role": "user",
                            "content": "Based on the description of the images above, what diagnostic conclusions can you draw?"
                        })
                message.append({
                            "role": "assistant",
                            "content": dp["image_Diagnosis"]
                        })
            all_data.append(messages)
```
